## E commerce Sales Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
```

```python
csv =pd.read_excel(r"C:\Users\lenovo\Desktop\Data set\Superstore.xlsx")
print(csv)
```

```python
csv.isnull().sum()
```

```python
des = csv.describe()
print(des)
```

```python
## Sort values
csv.sort_values(['Profit','Sales'],axis=0,ascending=[False,False],inplace=True)
```

```python
### graph b/w Region and Sales with Category
fig = px.bar(csv.head(20),x='Region',y='Sales',color='Category',width=1000)
fig.show()
```

```python
### Graph  b/w region and Sales with Profit
fig = px.bar(csv.head(20),x='Region',y='Sales',color='Profit',hover_data=['Category'],width=1000,title='Graph  b/w region and Sales with Profit')
fig.show()
```

```python
### graph order count v/s Sub Category
fig = px.bar(csv.head(100),x='Sub-Category',color='Category',width=700,title='Graph b/w order count and Sub-Category')
fig.show()
```

```python
### graph b/w Customer Name and Quantity with Profit
fig = px.bar(csv.head(20),x='Customer Name',y='Quantity',color='Profit',hover_data=['Discount','Category']
             ,width=700,height=600,title='Graph b/w Customer Name and Quantity with Profit')
fig.show()
```

```python
### Graph b/w Sub Category and Profit
fig = px.pie(csv,names='Sub-Category',values='Profit',title='Graph b/w Sub Category and Profit')
fig.show()
```
